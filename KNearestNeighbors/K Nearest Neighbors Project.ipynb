# K Nearest Neighbors Project 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Get the Data
** Read the 'KNN_Project_Data csv file into a dataframe **

```python
df = pd.read_csv('KNN_Project_Data')
```

```python
df.head()
```

# EDA


```python
sns.set_style('whitegrid')
sns.pairplot(data = df,hue='TARGET CLASS',palette='coolwarm')
```

# Standardize the Variables


```python
from sklearn.preprocessing import StandardScaler
```

** Create a StandardScaler() object called scaler.**

```python
scaler = StandardScaler()
```

** Fit scaler to the features.**

```python
scaler.fit(df.drop('TARGET CLASS',axis = 1))
```

**Use the .transform() method to transform the features to a scaled version.**

```python
scaled_feature = scaler.transform(df.drop('TARGET CLASS',axis = 1))
```

**Convert the scaled features to a dataframe and check the head of this dataframe to make sure the scaling worked.**

```python
df_features = pd.DataFrame(scaled_feature,columns = df.columns[:-1])
```

```python
df_features.head()
```

# Train Test Split

**Use train_test_split to split the data into a training set and a testing set.**

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_features
y= df['TARGET CLASS']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

# Using KNN


```python
from sklearn.neighbors import KNeighborsClassifier
```

**Create a KNN model instance with n_neighbors=1**

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

**Fit this KNN model to the training data.**

```python
knn.fit(X_train,y_train)
```

# Predictions and Evaluations


**Use the predict method to predict values using your KNN model and X_test.**

```python
predictions = knn.predict(X_test)
```

** Create a confusion matrix and classification report.**

```python
from sklearn.metrics import confusion_matrix, classification_report
```

** As we can see 72% accuracy might not be that good, so I decided to try other K value **

```python
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
```

# Choosing a K Value


** Create a for loop that trains various KNN models with different k values, then keep track of the error_rate for each of these models with a list.**

```python
error_rate = []

for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    predictions_i = knn.predict(X_test)
    error_rate.append(np.mean(predictions_i != y_test))
```

** Based on the graph we can see with the increase of the K value, the error rate start to drop, and it get pretty decent accuracy when K value is above 30 **

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

## Retrain with new K Value

**Retrain the model with the best K value and re-do the classification report and the confusion matrix. In my case it's 31. After that we can see the accuracy increased from 72 to 84. **

```python
knn = KNeighborsClassifier(n_neighbors=31)
knn.fit(X_train,y_train)
predictions = knn.predict(X_test)
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
```
